```python
# RUN THIS CODE BEFORE YOU START
import numpy as np
from matplotlib.pylab import plt
%matplotlib inline
```

```python
X = np.random.normal(0, 0.5,1000)
Y = np.random.normal(0, 2,1000)
Z = np.random.normal(2, 0.5,1000)
```

# Task 1: Histogram

Plot X in histogram, set the number of bins to 50.

**Hints: ** 
- `plt.hist`


```python
n_bins = 50
plt.hist(X, n_bins)
plt.show()
```

```python
plt.hist(X, 20, histtype='bar')
plt.hist(Y, 100, histtype='bar')
plt.hist(Z, 20, histtype='bar')
plt.show()
```

# Task 2: Multiple Histogram

Plot X, Y, Z on the same plot.

**Hints: **
- For visibility, set `histtyle` to "step" will be the best.

```python
plt.hist(X, 50, histtype='step')
plt.hist(Y, 200, histtype='step')
plt.hist(Z, 50, histtype='step')
plt.show()
```

# Task 3: Box Plot
Create box plot for X, Y, Z.

```python
plt.boxplot([X, Y, Z])
plt.show()
```

# Task 4: Quantile-Quantile Plot (qq plot)
Create qq plot for X.

```python
import scipy.stats as stats

stats.probplot(X, dist="norm", plot=plt)
plt.show()
```
