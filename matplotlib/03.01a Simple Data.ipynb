# Task 1: Plot a Simple Line
Plot $y = x^2,  x \in [-3, 3]$, the result should resumble the following:



**Hints **

# Task 3: Do Scatter Plot for Line with Error
Plot $y = x^2 + \epsilon,  x \in [-3, 3]$, the result should resumble the following:


**Hints **
- To introduce error, I use `np.random.normal` and a standard deviation of `0.4`.
- You will need to change the style using style character.

````python
X = np.linspace(-3, 3, 100)
error = np.random.normal(0,0.4,100) # Make an array of 100 numbers, with mean = 0, std = 0.4
Y = np.power(X, 2) + error
plt.plot(X, Y, 'r.') # r. means red dots
plt.show()
````
