````python
# RUN THIS CODE BEFORE YOU START
import numpy as np
from matplotlib.pylab import plt
%matplotlib inline
````

# Task 1: Plot Multiple Lines

I have prepared some random data, put all of them in one plot. The result should resemble the following:



**Hints: **Imagine a white canvas, each call to `plt.plot` will add a new line on the empty canvas.


````python
def create_random_walk():
    x = np.random.choice([-1,1],size=100, replace=True) # Sample with replacement from (-1, 1)
    return np.cumsum(x) # Return the cumulative sum of the elements
X = create_random_walk()
Y = create_random_walk()
Z = create_random_walk()
````

````python
plt.plot(X)
plt.plot(Y)
plt.plot(Z)
plt.show()
````

# Task 2: Change Line Style

The result should resemble the following:



**Hints: ** plt.plot(x, y, color="blue") will change the color to blue. For more color, refer to:  [http://matplotlib.org/users/colors.html](http://matplotlib.org/users/colors.html)

````python
plt.plot(X, '-.', color="#333333")
plt.plot(Y, '-.', color="chocolate")
plt.plot(Z, '-.', color="green")
plt.show()
````
