# Task 1: Show Image

I have downloaded a picture. Plot the image and increase the size so that you can see it clearly.

```python
import matplotlib.image as mpimg
url="https://cdn.pixabay.com/photo/2016/07/01/09/05/space-1490668_960_720.png"
img = mpimg.imread(url)
```

```python
fig = plt.figure(figsize=(13,13))
ax = fig.add_subplot(1,1,1)
ax.imshow(img)
```

# Task 2: Show 1 Channel of Image

If you look at the shape of img, you will find `(637, 960, 4)`. This means the image is 637 in height, 960 in width and there are 4 channels, i.e. RGBA (red, green, blue and alpha). Plot the blue (3rd) channel only, and set the `cmap` to "coolwarm". This will help you to see the key info in the picture.

```python
fig = plt.figure(figsize=(13,13))
ax = fig.add_subplot(1,1,1)
ax.imshow(img[:,:,2], cmap="coolwarm")
img.shape
```
